```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
url
```

```python
#response of page
response=requests.get(url)
response
```

```python
soup=BeautifulSoup(response.text,'html.parser')
```

```python
soup.prettify()
```

```python
#find table html code
table1=soup.find_all('table')[0].tbody
```

```python
#table1
```

```python
#find the rows from table
rows=table1.find_all('tr')
```

```python
#rows
```

```python
#extract column 1st row content title of column row[0] in th=table header
columns=[i.text.strip() for i in rows[0].find_all('th')]
columns
```

```python
#drop rank
columns.pop(0)
columns
```

#rank can extract from different  "th" tag

```python
r= []
# Loop through the rows  extract rank values from <th> tags
for j in range(2, len(rows)):  # Assuming rows[2:] contain the relevant data
    rank_data = rows[j].find('th').text.strip()
    r.append(rank_data)
```

```python
df1=pd.DataFrame(r,columns=['Rank'])
df1.head()
```

#other column can extract from "td" tag

```python
#extract other row from different tag "td"
d=[]
for i in range(2,len(rows)):
    data=rows[i].find_all('td')
    row=[tr.text.strip() for tr in data]
    d.append(row)
```

```python
#insert d in df
df2=pd.DataFrame(d,columns=columns)
df2.head()
```

#concat two DataFrame in to one

```python
#concat df1,df2
df3=pd.concat([df1,df2],axis=1)
df3.head()
```

#sate-owned column null value from "td" Tags
- so we drop it then again extract from other Tags

```python
#drop ref column unnessary columns
df3.drop('Ref.',axis=1,inplace=True)
df3.drop('State-owned',axis=1,inplace=True)
df3.head()
```

#State-owned is extracted from other  tag title

```python
# Find all td elements
td_elements = soup.find_all('td')
```

```python
# Extract the data-sort-value attribute expect from  corresponding text
values = []
for td in td_elements:
    data_sort_value = td.get('data-sort-value')
    if data_sort_value:  # Check if data-sort-value exists
        values.append(data_sort_value)

print(values)
```

```python
#create new df
df4=pd.DataFrame(values,columns=['State-owned'])
df4.head()
```

#now concat all dataset as ingle df

```python
df=pd.concat([df3,df4],axis=1)
df.head()
```

#save the extracted data in CSV file

```python
df.to_csv('Largest_company_by_revenue.csv',index=False)
```
